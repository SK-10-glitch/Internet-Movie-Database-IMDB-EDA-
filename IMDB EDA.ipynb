## <font color= 'Aqua'>DATA CLEANING AND DATA PREPARATION</font >

##### <font color = 'yellow'>Number of null values

```python
data.isnull().sum()
```

```python
data = data.rename(columns={"Series_Title": "Movies_Title"})
```

```python
data.head(2)
```

```python
print(data['Gross'].head(1))
data['Gross'] = data['Gross'].str.replace(',', '')
print(data['Gross'].head(1))

data['Gross'] = data['Gross'].astype('float64')
data['Gross'] = data['Gross'].replace(np.nan, 0)
```

```python
data['Gross'] = data['Gross'].astype(int)
```

```python
data['Gross'].dtype
```

## <FONT COLOR='Aqua'>EXPLORATORY ANALYSIS AND VISUALIZATION

## <font color = 'Gold'>Finding Correlation

```python
numerical_attributes = ['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']
data[numerical_attributes].hist(figsize = (15, 6), color = 'blue', edgecolor = 'red', layout = (2, 2))
```

```python
data.describe(include = ['O'])
```

## <font color = 'Lightsalmon'> Top Voted Movies

```python
top_voted = data.sort_values(['No_of_Votes'], ascending = False)
```

```python
fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_voted['Movies_Title'][:7],y=top_voted['No_of_Votes'][:7], palette = 'hls')
g.set_title("Top Voted Movies", weight = "bold")
plt.show()
```

## <font color = 'PeachPuff'> Top Voted Movies Ratings

```python
fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_voted['Movies_Title'][:7],y=top_voted['IMDB_Rating'][:7], palette = 'husl')
g.set_title("IMDB Rating of top voted movies", weight = "bold")
plt.show()
```

## <font color = 'DarkSeaGreen'> Top Voted Movies Metascores

```python
fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_voted['Movies_Title'][:7],y=top_voted['Meta_score'][:7], palette = 'husl')
g.set_title("Metascore of top rated movies", weight = "bold")
plt.show()
```

## <font color = 'Maroon'> Top Movies by Gross

```python
highest_earning = data.sort_values(['Gross'], ascending = False)
fig,axs=plt.subplots(figsize=(23,5))
g=sns.barplot(x=highest_earning['Movies_Title'][:7],y=highest_earning['Gross'][:7], palette = 'husl')
g.set_title("Movies with highest Gross", weight = "bold")
plt.show()
```

## <font color = 'Teal'> Maximum Movies Released

```python
fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(x=data['Released_Year'].value_counts()[:10].index,y=data['Released_Year'].value_counts()[:10])
g.set_title("Maximum Series released in-", weight = "bold")
g.set_xlabel("Years")
plt.show()
```

## <font color = 'LawnGreen'>Mostly Occurred Directors (Lineplot)

```python
fig,axs=plt.subplots(figsize=(20,5))
g=sns.lineplot(x=data['Director'].value_counts()[:10].index,y=data['Released_Year'].value_counts()[:10])
g.set_title("Mostly Occurred Directors", weight = "bold")
g.set_xlabel("Directors", weight = "bold")
plt.show()
```

## <font color = 'DarkViolet'>Mostly Occurred Directors (Barplot)

```python
fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(x=data['Director'].value_counts()[:10].index,y=data['Director'].value_counts()[:10])
g.set_title("Mostly Occurred Directors", weight = "bold")
g.set_xlabel("Directors", weight = "bold")
plt.show()
```

## <font color = 'Fuchsia'>Director and Stars Analysis

```python
top_directors = data['Director'].value_counts().head(10)
top_actors = data[['Star1', 'Star2', 'Star3', 'Star4']].stack().value_counts().head(10)

plt.bar(top_directors.index, top_directors.values)
plt.xticks(rotation=45)
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top Directors')
plt.show()

plt.bar(top_actors.index, top_actors.values)
plt.xticks(rotation=45)
plt.xlabel('Actor/Actress')
plt.ylabel('Count')
plt.title('Top Actors/Actresses')
plt.show()
```

## <font color = 'Lavender'>Directors with Respect to Gross of the Movies

```python
fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(x=data['Director'].value_counts()[:10].index,y=data['Gross'].value_counts()[:10])
g.set_title("Directors", weight = "bold")
g.set_xlabel("Mostly Occurred", weight = "bold")
plt.show()
```

## <font color = 'DarkKhaki'>IMDB Rating Distribution

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

sns.set(style="darkgrid")
g = sns.displot(data['IMDB_Rating'], bins=30, kde=True)
g.fig.suptitle("Distribution of Ratings", weight="bold")
plt.show()

```

##   <font color = 'MediumVioletRed'>Genre Analysis

```python
genre_counts = data['Genre'].value_counts()
top_genres = genre_counts.head(5)

plt.bar(top_genres.index, top_genres.values)
plt.xticks(rotation=45)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Top 5 Genres')
plt.show()
```

## <font color = 'OrangeRed'> Runtime Analysis </font>

```python
sns.histplot(data['Runtime'], bins=20)
plt.xlabel('Runtime')
plt.ylabel('Frequency')
plt.title('Runtime Distribution')
plt.show()
```

## <font color = 'DeepSkyBlue'>Certificate and IMDB Rating

```python
sns.boxplot(data=data, x='Certificate', y='IMDB_Rating')
plt.xticks(rotation=45)
plt.xlabel('Certificate')
plt.ylabel('IMDB Rating')
plt.title('IMDB Rating by Certificate')
plt.show()
```
